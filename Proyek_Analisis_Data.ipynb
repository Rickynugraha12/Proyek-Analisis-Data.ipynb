### Cleaning Data

**Insight:**
- Pada Tabel dataset `day_df` : Data ini memiliki kolom dteday yang sekarang bertipe datetime64[ns], yang memungkinkan analisis lebih lanjut berdasarkan tanggal untuk melihat pola musiman, mingguan, atau harian dalam jumlah penyewa sepeda (cnt). Dengan informasi tanggal yang terstruktur, kita bisa melakukan analisis tren penyewa sepeda berdasarkan musim, hari kerja, atau cuaca secara lebih mendalam.
- Pada Tabel dataset `hour_df` : Dengan memperbaiki tipe data kolom dteday menjadi datetime64[ns], analisis berbasis waktu menjadi lebih fleksibel dan efisien. Sekarang kita dapat melakukan manipulasi waktu yang lebih mudah, seperti ekstraksi informasi dari tanggal (misalnya, hari, bulan, atau tahun) dan melakukan analisis trend penyewaan sepeda yang lebih mendalam berdasarkan waktu (misalnya, penyewaan sepeda pada jam-jam tertentu atau berdasarkan musim).

#### Membersihkan tabel `day_df`

```python
day_df.info()
```

##### Memperbaiki Tipe Data

```python
datetime_columns = ["dteday"]

for column in datetime_columns:
  day_df[column] = pd.to_datetime(day_df[column])
```

```python
day_df.head()
```

##### Memperbaiki Tipe Data

```python
datetime_columns = ["dteday"]

for column in datetime_columns:
  hour_df[column] = pd.to_datetime(hour_df[column])
```

### Explore `day_df`

```python
day_df.sample(5)
```

```python
day_df.describe(include="all")
```

```python
day_df.groupby(by="weathersit").agg({
    "instant": "nunique",
    "season": ["max", "min", "mean", "std", ]
})
```

```python
day_df.groupby(by="weathersit").instant.nunique().sort_values(ascending=False)
```

### Explore `hour_df`

```python
hour_df.sample(5)
```

```python
hour_df.describe(include="all")
```

```python
hour_df.groupby(by="hr").agg({
    "instant": "nunique",
    "cnt": ["max", "min", "mean", "std", ]
})
```

```python
hour_df.groupby(by="hr").instant.nunique().sort_values(ascending=False)
```

```python
hour_df.groupby(by="cnt").instant.nunique().sort_values(ascending=False)
```

### Explore all data

#### Menghubungkan semua data

```python
all_df = pd.merge(
    left=day_df,
    right=hour_df,
    how="left",
    left_on="instant",
    right_on="instant"
)
all_df.head()
```

```python
all_df.groupby(by=["instant"]).agg({
    "season_x": "max",
    "season_y": "min",
    "cnt_x": "max",
    "cnt_y": "min"
})
```

```python
all_df.groupby(by=["instant"]).agg({
    "season_x": "max",
    "season_y": "min",
    "cnt_x": "max",
    "cnt_y": "min"
}).sort_values(by="instant", ascending=False)
```
